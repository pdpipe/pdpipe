```python
from random import randint
```

```python
data = [
    [randint(1,10), v]
    for v in ind_vals
]
data
```

```python
df = pd.DataFrame(data, index=range(len(ind_vals)), columns=["a", "Industria"])
df['Industria'] = df['Industria'].astype('category')
df
```

```python
pline = pdp.PdPipeline([
    pdp.OneHotEncode(columns=['Industria']),
])
```

```python
pline
```

```python
post_ft = pline.fit_transform(df)
```

```python
post_ft
```
